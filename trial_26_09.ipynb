```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
audio_path = 'noise_1.wav'
audio, sr = librosa.load(audio_path, sr=16000) 
```

```python
audio = librosa.to_mono(audio)
```

```python
stft_result = librosa.stft(audio, n_fft=2048, hop_length=512)
magnitude = np.abs(stft_result)
```

```python
spectrogram_db = librosa.amplitude_to_db(magnitude, ref=np.max)
```

```python
plt.figure(figsize=(10, 6))
librosa.display.specshow(spectrogram_db, sr=sr, hop_length=512, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('STFT Spectrogram (dB)')
plt.show()
```

```python
fft_result = np.fft.fft(audio)

# Compute frequencies associated with FFT result
frequencies = np.fft.fftfreq(len(fft_result), 1/sr)

# Only use the positive frequencies (FFT result is symmetric for real inputs)
positive_freqs = frequencies[:len(frequencies)//2]
magnitude = np.abs(fft_result[:len(fft_result)//2])

# Plot the FFT (Magnitude vs Frequency)
plt.figure(figsize=(10, 6))
plt.plot(positive_freqs, magnitude)
plt.title('FFT of Audio Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```
